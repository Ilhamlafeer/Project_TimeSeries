```python
# TIME BASED FEATURES
df_feat = df.copy()

df_feat['hour'] = df_feat.index.hour
df_feat['day_of_week'] = df_feat.index.dayofweek
df_feat['month'] = df_feat.index.month

# Weekend or weekday
df_feat['is_weekend'] = df_feat['day_of_week'].isin([5,6]).astype(int)

# Night indicator (0:00 to 5:00)
df_feat['is_night'] = df_feat['hour'].isin([0,1,2,3,4,5]).astype(int)
```

```python
# MOVING WINDOW FEATURES
df_feat['roll_mean_1h'] = df_feat['Appliances'].rolling(6).mean()
df_feat['roll_std_1h']  = df_feat['Appliances'].rolling(6).std()

df_feat['roll_mean_3h'] = df_feat['Appliances'].rolling(18).mean()
df_feat['roll_std_3h']  = df_feat['Appliances'].rolling(18).std()
```

```python
# LAG FEATURES
for lag in [1, 6, 12, 24]:
    df_feat[f'lag_{lag}'] = df_feat['Appliances'].shift(lag)
```

```python
# INTERACTION FEATURES
# Average indoor temp and humidity
temp_cols = [c for c in df_feat.columns if c.startswith('T')]
rh_cols   = [c for c in df_feat.columns if c.startswith('RH')]

df_feat['T_avg'] = df_feat[temp_cols].mean(axis=1)
df_feat['RH_avg'] = df_feat[rh_cols].mean(axis=1)

# Temp-humidity interaction
df_feat['T_RH_interaction'] = df_feat['T_avg'] * df_feat['RH_avg']

# Lights × lag interaction
df_feat['lights_x_lag1'] = df_feat['lights'] * df_feat['lag_1']
```

```python
# DOMAIN-SPECIFIC FEATURES
df_feat['is_morning'] = df_feat['hour'].between(6, 11).astype(int)
df_feat['is_evening'] = df_feat['hour'].between(17, 23).astype(int)

# Hot months ---- high AC usage
df_feat['is_hot_season'] = df_feat['month'].isin([4,5,6,7,8,9]).astype(int)
```

```python
# FEATURE SELECTION
from sklearn.ensemble import RandomForestRegressor

df_fs = df_feat.dropna()

X = df_fs.drop(columns=['Appliances'])
y = df_fs['Appliances']

rf = RandomForestRegressor(n_estimators=120, random_state=42)
rf.fit(X, y)

importances = pd.Series(rf.feature_importances_, index=X.columns)
importances.sort_values(ascending=False).head(20)
```

```python
# Save processed dataset
df_processed = df_feat.dropna()
df_processed.to_csv("energy_data_processed.csv")
```

# TASK 3(B): Deep Learning Model Development

```python
# Load processed dataset
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/Internship/energy_data_processed.csv", parse_dates=['date'], index_col='date')
target = "Appliances"
```

```python
# Windowing Function (Universal for all DL models)
import numpy as np
def create_sequences(data, target_col, window_size=24):
    X = []
    y = []

    for i in range(window_size, len(data)):
        X.append(data.iloc[i-window_size:i].values)
        y.append(data.iloc[i][target_col])

    return np.array(X), np.array(y)
```

```python
# Train/Test Split (Time-Based)
split = int(len(df) * 0.8)
train_df = df.iloc[:split]
test_df  = df.iloc[split:]
```

```python
# Create LSTM-ready sequences
# Input features
feature_cols = df.columns.tolist()
feature_cols.remove(target)

X_train, y_train = create_sequences(train_df[feature_cols + [target]], target_col=target)
X_test, y_test   = create_sequences(test_df[feature_cols + [target]], target_col=target)

# Print shapes
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# Build LSTM Model
```

```python
# LSTM Architecture
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model_lstm = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),

    LSTM(32, return_sequences=False),
    Dropout(0.2),

    Dense(16, activation='relu'),
    Dense(1)
])

model_lstm.compile(
    loss='mse',
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=['mae']
)

model_lstm.summary()
```

```python
# Train LSTM Model
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history_lstm = model_lstm.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50,
    batch_size=32,
    callbacks=[es],
    verbose=1
)
```

```python
# Evaluate LSTM
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

pred_lstm = model_lstm.predict(X_test)
mae_lstm = mean_absolute_error(y_test, pred_lstm)
rmse_lstm = np.sqrt(mean_squared_error(y_test, pred_lstm))

print("LSTM MAE:", mae_lstm)
print("LSTM RMSE:", rmse_lstm)
```

# Build GRU Model

```python
# GRU Architecture
from tensorflow.keras.layers import GRU

model_gru = Sequential([
    GRU(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),

    GRU(32, return_sequences=False),
    Dropout(0.2),

    Dense(16, activation='relu'),
    Dense(1)
])

model_gru.compile(
    loss='mse',
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=['mae']
)
```

```python
# Train GRU
history_gru = model_gru.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50,
    batch_size=32,
    callbacks=[es],
    verbose=1
)
```

```python
# Evaluate GRU
pred_gru = model_gru.predict(X_test)
mae_gru = mean_absolute_error(y_test, pred_gru)
rmse_gru = np.sqrt(mean_squared_error(y_test, pred_gru))

print("GRU MAE:", mae_gru)
print("GRU RMSE:", rmse_gru)
```

# CNN-LSTM Hybrid model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout

model_cnn_lstm = Sequential([
    # CNN block
    Conv1D(filters=64, kernel_size=3, activation='relu',
           input_shape=(X_train.shape[1], X_train.shape[2])),
    MaxPooling1D(pool_size=2),

    # LSTM block
    LSTM(64),
    Dropout(0.2),

    # Dense block
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),

    # Output
    Dense(1)
])

model_cnn_lstm.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

model_cnn_lstm.summary()
```

```python
# Training
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

es = EarlyStopping(
    monitor='val_loss',
    patience=8,
    restore_best_weights=True
)

lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=4,
    min_lr=1e-6
)

history = model_cnn_lstm.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50,
    batch_size=32,
    callbacks=[es, lr],
    verbose=1
)
```

```python
pred_lstm = model_lstm.predict(X_test)
pred_gru = model_gru.predict(X_test)
pred_cnn_lstm = model_cnn_lstm.predict(X_test)
```

```python
# Compute Evaluation Metrics (MAE, RMSE, MAPE, R²)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, mape, r2

print("LSTM:", evaluate(y_test, pred_lstm))
print("GRU:", evaluate(y_test, pred_gru))
print("CNN-LSTM:", evaluate(y_test, pred_cnn_lstm))
```

```python
def create_sequences(df, window=48):
    X, y = [], []
    features = df.drop(columns=['Appliances']).values
    target = df['Appliances'].values

    for i in range(window, len(df)):
        X.append(features[i-window:i])
        y.append(target[i])
    return np.array(X), np.array(y)
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/Internship/energy_data_processed.csv", parse_dates=['date'], index_col='date')
split = int(len(df) * 0.8)
train_df = df.iloc[:split]
test_df  = df.iloc[split:]
```

```python
# Create 3D sequences
WINDOW = 24

X_train, y_train = create_sequences(train_df, WINDOW)
X_test, y_test   = create_sequences(test_df, WINDOW)

print(X_train.shape)
print(y_train.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model_lstm = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),

    LSTM(32),
    Dropout(0.2),

    Dense(16, activation='relu'),
    Dense(1)
])

model_lstm.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']
)

history_lstm = model_lstm.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50,
    batch_size=32,
    callbacks=[es, rlrop],
    verbose=1
)
```

```python
# Optimize GRU Model
from tensorflow.keras.layers import GRU

model_gru = Sequential([
    GRU(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),

    GRU(32),
    Dropout(0.2),

    Dense(32, activation='relu'),
    Dense(1)
])

model_gru.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']
)

history_gru = model_gru.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=60,
    batch_size=32,
    callbacks=[es, rlrop],
    verbose=1
)
```

```python
# Optimize CNN-LSTM
from tensorflow.keras.layers import Conv1D, MaxPooling1D

model_cnn_lstm = Sequential([
    # CNN feature extraction
    Conv1D(filters=64, kernel_size=3, activation='relu',
           input_shape=(X_train.shape[1], X_train.shape[2])),
    MaxPooling1D(pool_size=2),

    # LSTM layer
    LSTM(64),
    Dropout(0.2),

    # Dense layers
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),

    Dense(1)
])

model_cnn_lstm.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']
)

history_cnn_lstm = model_cnn_lstm.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=60,
    batch_size=32,
    callbacks=[es, rlrop],
    verbose=1
)
```
